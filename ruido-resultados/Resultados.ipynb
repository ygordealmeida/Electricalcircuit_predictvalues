```python
import pickle
import matplotlib.pyplot as plt
import scienceplots

with open('historico_150_10db', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    historico_150_10db = pickle.load(file)

with open('historico_150_20db', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    historico_150_20db = pickle.load(file)
with open('historico_150_30db', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    historico_150_30db = pickle.load(file)



loss10db = historico_150_10db['loss']
val_loss10db = historico_150_10db['val_loss']

loss20db = historico_150_20db['loss']
val_loss20db = historico_150_20db['val_loss']

loss30db = historico_150_30db['loss']
val_loss30db = historico_150_30db['val_loss']

epochs = range(1, len(loss10db) + 1)

# Plote os grficos
plt.figure()
#plt.figure(2, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.style.use(['science','ieee'])

plt.plot(epochs, val_loss10db, 'r', linewidth=3)
plt.plot(epochs, val_loss20db, 'g', linewidth=3)
plt.plot(epochs, val_loss30db, 'm', linewidth=3)
plt.xlim(1,30)

plt.xlabel('pocas',fontsize=20)

plt.ylabel('Erro de Validao', fontsize=20)
plt.legend(('Erro SNR = 10dB','Erro SNR = 20dB','Erro SNR = 30dB'), prop={'size': 12.5})
plt.grid()
plt.yscale("log")

plt.tick_params(axis='both', which='major', labelsize=18)  # Tamanho dos rtulos principais
plt.tick_params(axis='both', which='minor', labelsize=18)  # Tamanho dos rtulos menores

plt.tight_layout()
plt.show()
```
