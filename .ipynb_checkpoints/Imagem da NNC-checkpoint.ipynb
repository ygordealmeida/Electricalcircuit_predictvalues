```python
import tensorflow as tf 
from tensorflow.keras import Sequential, regularizers
from tensorflow.keras.layers import Dense,Dropout, Conv2D, MaxPooling2D, Flatten


#Esse Ã© a arquitetura da nossa rede neural:
model_2 = Sequential ([
    Conv2D(64, (3,3), activation='tanh', input_shape=(200,200,1)),
    MaxPooling2D(2,2),  #MAX POOLING AGRUPA OS PIXELS DA IMAGEM E OS FILTRA EM UM SUBCONJUNTO
    Conv2D(64, (3,3),activation='tanh'),
    MaxPooling2D(2,2),
    Conv2D(128,(3,3),activation='tanh'),
    MaxPooling2D(2,2),
    Conv2D(128,(3,3),activation='tanh'),
    MaxPooling2D(2,2),
    Conv2D(256,(3,3),activation='tanh'),
    MaxPooling2D(2,2),
    Flatten(),
    Dropout(0.5),
    Dense(512,activation='tanh'),
    Dense (units=3, activation='linear')
    
])


from tensorflow.keras.losses import MeanSquaredError, MeanSquaredLogarithmicError 

model_2.compile (optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
loss=MeanSquaredLogarithmicError(), 
metrics=['accuracy']) 

```
