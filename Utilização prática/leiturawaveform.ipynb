```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Ler o arquivo txt
data = pd.read_csv('aquisicao04.txt', sep='\t')

# Extrair as colunas em variáveis separadas

time = data['Time (s)'].values
#channel_1 é a tensão no resistor do circuito RLC série
channel_1 = data['Channel 1 (V)'].values

channel_1_corrente= channel_1/300 #Divide pela resistência de 300 ohms
# Mostrar os dados
print('Time (s):', time)
print('Channel 1 (V):', channel_1)
```

```python
plt.plot(time,channel_1)
plt.title("Leitura do osciloscopio")
plt.ylabel("Tensão no resistor (V)")
plt.xlabel("Tempo (s)")
```

```python
plt.plot(time,channel_1_corrente)
plt.title("Gráfico corrente em função do tempo")
plt.ylabel("Corrente elétrica (A)")
plt.xlabel("Tempo (s)")

plt.show()
```

```python
#Salva a figura de i(t) no formato apropriado para a rede neural

plt.figure(figsize=(10,7), dpi=100, facecolor='w', edgecolor='k')
plt.plot(time,channel_1_corrente,color='black')
plt.grid(False)
plt.xlim(-1e-6 -1.4745454545454549e-06,-1e-6 -1.4745454545454549e-06 + 1/10000)
plt.xticks([])
y_ticks = np.arange(-0.04, 0.03, 0.01)  # Divisões no eixo y
plt.yticks(y_ticks)
plt.gca().set_yticklabels(['' for _ in y_ticks])
plt.savefig("aquisicao04.jpg", format='jpg', bbox_inches='tight')

plt.show(block=False)
```
