```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
# Abrir o arquivo para leitura binária ('rb')
with open('historico_200', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    loaded_history = pickle.load(file)

# Agora, 'loaded_history' contém os dados carregados do arquivo
#print(loaded_history)

# Extraia os dados do histórico
acc = loaded_history['accuracy']
val_acc = loaded_history['val_accuracy']
loss = loaded_history['loss']
val_loss = loaded_history['val_loss']
epochs = range(1, len(acc) + 1)

# Plote os gráficos
plt.plot(epochs, loss, 'r', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
