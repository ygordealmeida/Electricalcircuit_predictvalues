```python
import pickle
import matplotlib.pyplot as plt
import scienceplots

with open('historico_200', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    loaded_history200 = pickle.load(file)

with open('historico_150', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    loaded_history150 = pickle.load(file)
with open('historico_125', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    loaded_history125 = pickle.load(file)
with open('historico_100', 'rb') as file:
    # Carregar os dados do arquivo usando pickle
    loaded_history100 = pickle.load(file)


loss200 = loaded_history200['loss']
val_loss200 = loaded_history200['val_loss']

loss150 = loaded_history150['loss']
val_loss150 = loaded_history150['val_loss']

loss125 = loaded_history125['loss']
val_loss125 = loaded_history125['val_loss']

loss100 = loaded_history100['loss']
val_loss100 = loaded_history100['val_loss']

epochs = range(1, len(loss200) + 1)

# Plote os gráficos
plt.figure()
#plt.figure(2, figsize=(8, 6), dpi=100, facecolor='w', edgecolor='k')
#plt.style.use('science')
plt.style.use(['science','ieee'])
plt.plot(epochs, val_loss200, 'r',linewidth=3,)
plt.plot(epochs, val_loss150, 'g',linewidth=3)
plt.plot(epochs, val_loss125, 'm',linewidth=3)
plt.plot(epochs, val_loss100, 'k',linewidth=3)

plt.xlabel('Épocas',fontsize=20)
plt.xlim(1,30)


plt.ylabel('Erro de Validação', fontsize=20)
plt.legend(('Erro 200px','Erro 150px','Erro 125px','Erro 100px'), prop={'size': 12.5})
plt.grid()
plt.yscale("log")

plt.tick_params(axis='both', which='major', labelsize=18)  # Tamanho dos rótulos principais
plt.tick_params(axis='both', which='minor', labelsize=18)  # Tamanho dos rótulos menores

plt.tight_layout()
plt.show()
```
